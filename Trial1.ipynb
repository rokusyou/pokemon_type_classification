### Pattern 1

```python
# Create PK Skill Dictionary

skill_dic ={}
skill_num = 1
TOT_PK_NUM = 800

# PK iter
for i in range(1,TOT_PK_NUM+1):
    
    # skill iter
    for skill in pk_zkan[str(i)]['skills']:
        
        skill_regist = skill in skill_dic.values()
        #print( "skill name = %s , Dic_exist = %s" % (skill, skill_regist ) )
        if skill_regist ==False:
            skill_dic[skill_num] = skill
            skill_num +=1


            

# Confirm PK skill Dictionary
skill_dic
```
